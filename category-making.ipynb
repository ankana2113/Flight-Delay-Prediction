```python
def category_make(df, df1):
    df1['CRSDepHours'] = np.ceil(df['CRSDepHours']/3).apply(int)
    df1['DepHours'] = np.ceil(df['DepHours']/3).apply(int)
    df1['CRSArrHours'] = np.ceil(df['CRSArrHours']/3).apply(int)
    df1['ArrHours'] = np.ceil(df['ArrHours']/3).apply(int)
  
    df1['CRSDepMinutes'] = np.ceil(df['CRSDepMinutes']/15).apply(int)
    df1['DepMinutes'] = np.ceil(df['DepMinutes']/15).apply(int)
    df1['CRSArrMinutes'] = np.ceil(df['CRSArrMinutes']/15).apply(int)
    df1['ArrMinutes'] = np.ceil(df['ArrMinutes']/15).apply(int)

    df1['DepDelayMinutes'] = np.ceil(df['DepDelayMinutes']/30).apply(int)
    df1['ArrDelayMinutes'] = np.ceil(df['ArrDelayMinutes']/30).apply(int)
    df1['CRSElapsedTime'] = np.ceil(df['CRSElapsedTime']/30).apply(int)
    df1['ActualElapsedTime'] = np.ceil(df['ActualElapsedTime']/30).apply(int)
    
    df1['AirTime'] = np.ceil(df['AirTime']/15).apply(int)
    df1['Distance'] = np.ceil((df['Distance']-30)/50).apply(int)

    df1['CarrierDelay'] = np.ceil(df['CarrierDelay']/15).apply(int)
    df1['WeatherDelay'] = np.ceil(df['WeatherDelay']/15).apply(int)
    df1['NASDelay'] = np.ceil(df['NASDelay']/900).apply(int)
    df1['SecurityDelay'] = np.ceil(df['SecurityDelay']/15).apply(int)
    df1['LateAircraftDelay'] = np.ceil(df['LateAircraftDelay']/15).apply(int)
    
    df1['tavg'] = np.ceil((df['tavg']-min(df['tavg']))/10).apply(int)
    df1['tmin'] = np.ceil((df['tmin']-min(df['tmin']))/10).apply(int)
    df1['tmax'] = np.ceil((df['tmax']-min(df['tmax']))/10).apply(int)
    df1['prcp'] = np.ceil(df['prcp']/50).apply(int)
    df1['wdir'] = np.ceil(df['wdir']/30).apply(int)
    df1['wspd'] = np.ceil(df['wspd']/10).apply(int)
    df1['pres'] = np.ceil((df['pres']-min(df['pres']))/10).apply(int)

    return df1
```

```python
train_df1 = category_make(train_df1, train_df1)
# train_df1.head()
```
