```python
cancellation_code_mapping = {
    np.nan : 0,
    'A' : 1,
    'B' : 2,
    'C' : 3,
    'D' : 4,
}

df1['CancellationCode_encoded'] = df1['CancellationCode'].map(cancellation_code_mapping).fillna(0).astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1['Z_Score'] = (df1['ArrDelayMinutes'] - df1['ArrDelayMinutes'].mean()) / df1['ArrDelayMinutes'].std()

# Flag outliers
outliers_zscore = df1[np.abs(df1['Z_Score']) > 3]
# print(outliers.T)
print(len(outliers_zscore))
print(df1.shape)
```

```python
df1.shape
```

```python
Q1 = df1['ArrDelayMinutes'].quantile(0.25)
Q3 = df1['ArrDelayMinutes'].quantile(0.75)
IQR = Q3 - Q1

# Flag outliers
outliers_quantile = df1[(df1['ArrDelayMinutes'] < (Q1 - 1.5 * IQR)) | (df1['ArrDelayMinutes'] > (Q3 + 1.5 * IQR))]
print(outliers_quantile.shape[0])
print(df1.shape)
```

```python
df1.fillna(0, inplace=True)

from sklearn.ensemble import IsolationForest

# Instantiate Isolation Forest
iso_forest = IsolationForest(contamination=0.1)  # Set contamination rate
df1['Outlier_iso'] = iso_forest.fit_predict(df1[['ArrDelayMinutes']])

# Extract outliers (outliers will have -1, inliers will have 1)
outliers_iso = df1[df1['Outlier_iso'] == -1]
print(outliers_iso.shape[0])
print(df1.shape)
```

```python
# Filter the DataFrame to keep rows where ArrDelay15 is not equal to 0
non_delay_outliers_zscore = outliers_zscore[outliers_zscore['ArrDel15'] == 0.0]
non_delay_outliers_quantile= outliers_quantile[outliers_quantile['ArrDel15'] == 0.0]
non_delay_outliers_iso = outliers_iso[outliers_iso['ArrDel15'] == 0.0]

# Display the new DataFrame with non-zero ArrDelay15 values
print(non_delay_outliers_zscore)
print(non_delay_outliers_quantile)
print(non_delay_outliers_iso)
print(df1.shape)
```
