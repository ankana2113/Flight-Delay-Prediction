```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('/kaggle/input/airflight-final/train_categorised.csv')
test_df = pd.read_csv('/kaggle/input/airflight-final/test_categorised.csv')
```

```python
train_df.head().T
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import accuracy_score
```

```python
train_df.columns
```

```python
x_train = train_df.drop(['DepDelayMinutes', 'ArrDelayMinutes'], axis=1)
y_train = train_df[['DepDelayMinutes', 'ArrDelayMinutes']]

x_test = test_df.drop(['DepDelayMinutes', 'ArrDelayMinutes'], axis=1)
y_test = test_df[['DepDelayMinutes', 'ArrDelayMinutes']]
```

```python
# Define base models
base_estimators = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state=42)),
    ('svc', SVC(probability=True, random_state=42))
]

# Define the stacking classifier
stacking_clf = StackingClassifier(
    estimators=base_estimators,       # Base models
    final_estimator=LogisticRegression()  # Meta-model
)

# Use MultiOutputClassifier to handle multi-output case
multioutput_clf = MultiOutputClassifier(stacking_clf)
```

```python
# Fit the multi-output stacking classifier
multioutput_clf.fit(x_train, y_train)
```

```python
# Make predictions and evaluate
y_pred = multioutput_clf.predict(x_test)

# Calculate accuracy for each output
accuracy = [accuracy_score(y_test[:, i], y_pred[:, i]) for i in range(y_train.shape[1])]
print("Stacking Classifier Accuracy for each output:", accuracy)
```
