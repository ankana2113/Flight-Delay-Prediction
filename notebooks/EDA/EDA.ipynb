```python
import pandas as pd
import numpy as np
# import pandas_profiling
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("/content/final.csv")
df.head()
```

```python
# Filter for the year 2024
df = df[df['Year'] == 2024]

# Convert FlightDate to datetime format
df['FlightDate'] = pd.to_datetime(df['FlightDate'], format='%d-%m-%Y',errors='coerce')

# Extract the month from FlightDate
df['Month'] = df['FlightDate'].dt.month

# Group by month and calculate the average delay
monthly_delay = df.groupby('Month')['DepDelayMinutes'].mean().reset_index()

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='DepDelayMinutes', data=monthly_delay, palette='viridis')

# Customize the plot
plt.title('Average Departure Delay per Month in 2024')
plt.xlabel('Month')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y')

# Show the plot
plt.show()
```
