```python
import pandas as pd
df=pd.read_csv('/kaggle/input/pp-data/final_preprocessed_data.csv',low_memory=False)
```

```python

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf




# Select only the numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Set the number of lags for ACF and PACF
lags = 30  

# Plot ACF and PACF for each numerical column
for column in numerical_df.columns:
    print(f"\nACF and PACF for {column}")
    
    # Create subplots for ACF and PACF
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))
    fig.suptitle(f'ACF and PACF for {column}')
    
    # Plot ACF
    plot_acf(numerical_df[column].dropna(), lags=lags, ax=axes[0], title=f"ACF of {column}")
    
    # Plot PACF
    plot_pacf(numerical_df[column].dropna(), lags=lags, ax=axes[1], title=f"PACF of {column}")
    
    # Show the plots
    plt.show()
```

```python
df['FlightDate'] = pd.to_datetime(df['FlightDate'], errors='coerce', format='%m/%d/%Y')

# Step 2: Handle alternative format without inplace
df['FlightDate'] = df['FlightDate'].fillna(pd.to_datetime(df['FlightDate'], errors='coerce', format='%d-%m-%Y'))

# Step 3: Sort by the date column
df = df.sort_values(by='FlightDate').reset_index(drop=True)
print(df)
```
